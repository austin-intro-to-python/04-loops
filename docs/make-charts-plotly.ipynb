## Generate HTML images

```python
import plotly.graph_objs as go
```

Figure 1. Bar Chart

```python
### Create some simple data
mylabels = ['Breakfast','Lunch','Dinner','Snacks']
myvalues = [10,11,2,6]
```

```python
# Combine those into a bar chart
mydata = go.Bar(x=mylabels,  y=myvalues)
fig = go.Figure([mydata])
```

```python
# save the figure as html
fig.write_html("scatterplot.html")
```

```python
########### Define your variables ######
mylabels = ['Chocolate Frosted', 'French Cruller', 'Boston Creme']
myvalues = [45,25,7]
color1 = 'e11383'
color2 = 'f5821f'
color3 = '683817'
```

Figure 2. Donut Chart

```python
########### Set up the data for the chart
mydata = go.Pie(
    values=myvalues,
    labels=mylabels,
    hole=0.5,
    marker={'colors': [color1, color2, color3],
            'line': {'color': 'white', 'width': 5}}
)
mydata
```

```python
# Add some layout
mylayout = go.Layout(title = "What's your favorite Dunkin' Donut?")
mylayout
```

```python
# combine data and layout into a figure
fig = go.Figure(data=[mydata], layout=mylayout)
fig
```

```python
# save the figure as html
fig.write_html("piechart.html")
```
